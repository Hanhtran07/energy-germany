```python
df = df.rename(columns={
        'Start date': 'timestamp_start',
        'End date': 'timestamp_end',
        'Biomass [MWh] Original resolutions': 'biomass_mwh',
        'Hydropower [MWh] Original resolutions': 'hydro_mwh',
        'Wind offshore [MWh] Original resolutions': 'wind_offshore_mwh',
        'Wind onshore [MWh] Original resolutions': 'wind_onshore_mwh',
        'Photovoltaics [MWh] Original resolutions': 'solar_mwh',
        'Other renewable [MWh] Original resolutions': 'other_renewable_mwh',
        'Nuclear [MWh] Original resolutions': 'nuclear_mwh',
        'Lignite [MWh] Original resolutions': 'lignite_mwh',
        'Hard coal [MWh] Original resolutions': 'hard_coal_mwh',
        'Fossil gas [MWh] Original resolutions': 'gas_mwh',
        'Hydro pumped storage [MWh] Original resolutions': 'hydro_pumped_mwh',
        'Other conventional [MWh] Original resolutions': 'other_conventional_mwh'
    })
```

```python
print("Nuclear sau rename:")
print(df['nuclear_mwh'].head(3).tolist())
print(f"Nuclear dtype: {df['nuclear_mwh'].dtype}")
print(f"Nuclear mean raw: {pd.to_numeric(df['nuclear_mwh'], errors='coerce').mean():.2f}")
```

```python

df['nuclear_mwh'] = pd.to_numeric(
    df['nuclear_mwh'], errors='coerce'
).fillna(0)

df['other_conventional_mwh'] = pd.to_numeric(
    df['other_conventional_mwh'], errors='coerce'
).fillna(0)

#check
print(f"Nuclear mean sau fix: {df['nuclear_mwh'].mean():.2f}")
print(f"Nuclear 2018 sample: {df['nuclear_mwh'].head(5).tolist()}")
```

```python
#change date type
df['timestamp_start'] = pd.to_datetime(df['timestamp_start'])
df['timestamp_end']   = pd.to_datetime(df['timestamp_end'])
```

```python
# ── Extract time features ─────────────────────────────
df['year']    = df['timestamp_start'].dt.year
df['month']   = df['timestamp_start'].dt.month
df['day']     = df['timestamp_start'].dt.day
df['hour']    = df['timestamp_start'].dt.hour
df['weekday'] = df['timestamp_start'].dt.day_name()
df['season']  = df['month'].map({
    12:'Winter', 1:'Winter',  2:'Winter',
     3:'Spring', 4:'Spring',  5:'Spring',
     6:'Summer', 7:'Summer',  8:'Summer',
     9:'Autumn',10:'Autumn', 11:'Autumn'
})
```

```python
# ── Derived columns ───────────────────────────────────
df['renewable_mwh'] = (
    df['biomass_mwh'] + df['hydro_mwh'] +
    df['wind_offshore_mwh'] + df['wind_onshore_mwh'] +
    df['solar_mwh'] + df['other_renewable_mwh']
)
df['fossil_mwh'] = (
    df['lignite_mwh'] + df['hard_coal_mwh'] +
    df['gas_mwh'] + df['other_conventional_mwh']
)
df['total_mwh'] = (
    df['renewable_mwh'] + df['fossil_mwh'] + df['nuclear_mwh'] 
)
df['renewable_pct'] = (
    df['renewable_mwh'] / df['total_mwh'] * 100
).round(2)
```

```python
def classify_green_hour(pct):
    if pct >= 80:
        return 'Very Green'
    elif pct >= 50:
        return 'Green'
    else:
        return 'Not Green'

df['green_label'] = df['renewable_pct'].apply(classify_green_hour)
```

```python
# ── VERIFY ────────────────────────────────────────────
print(f"Hydro mean:      {df['hydro_mwh'].mean():.2f}")      # ~432
print(f"Avg renewable %: {df['renewable_pct'].mean():.1f}%") # ~47-48%
print(f"NaN check:       {df['renewable_pct'].isna().sum()}") # 0
print(f"\nYearly:")
print(df.groupby('year')['renewable_pct'].mean().round(1))
print(f"\nGreen label:")
print(df['green_label'].value_counts())
```

```python
# ── Save fct_summary ───
fct_summary = df[[
    'timestamp_start', 'year', 'month', 'day', 'hour',
    'weekday', 'season',
    'biomass_mwh', 'hydro_mwh', 'wind_offshore_mwh',
    'wind_onshore_mwh', 'solar_mwh', 'other_renewable_mwh',
    'nuclear_mwh', 'lignite_mwh', 'hard_coal_mwh',
    'gas_mwh', 'hydro_pumped_mwh', 'other_conventional_mwh',
    'renewable_mwh', 'fossil_mwh', 'total_mwh',
    'renewable_pct', 'green_label'
]].copy()
```

```python
# ── dim_time ──────────────────────────────────────────
dim_time = df[[
    'timestamp_start', 'year', 'month', 'day',
    'hour', 'weekday', 'season'
]].drop_duplicates().reset_index(drop=True)
dim_time.insert(0, 'time_id', range(1, len(dim_time) + 1))
```

```python
# ── dim_energy_source ─────────────────────────────────
dim_energy_source = pd.DataFrame({
    'source_id':   range(1, 13),
    'source_name': ['Biomass','Hydropower','Wind Offshore',
                    'Wind Onshore','Solar','Other Renewable',
                    'Nuclear','Lignite','Hard Coal',
                    'Fossil Gas','Other Conventional',  'Hydro Pumped Storage'],
    'source_type': ['Renewable','Renewable','Renewable',
                    'Renewable','Renewable','Renewable',
                    'Nuclear','Fossil','Fossil','Fossil','Fossil', 'Storage'],
    'is_renewable':[True,True,True,True,True,True,
                    False,False,False,False,False,False]
})
```

```python
# ── fct_generation (long format) ─────────────────────
source_col_map = {
    'biomass_mwh':           'Biomass',
    'hydro_mwh':             'Hydropower',
    'wind_offshore_mwh':     'Wind Offshore',
    'wind_onshore_mwh':      'Wind Onshore',
    'solar_mwh':             'Solar',
    'other_renewable_mwh':   'Other Renewable',
    'nuclear_mwh':           'Nuclear',
    'lignite_mwh':           'Lignite',
    'hard_coal_mwh':         'Hard Coal',
    'gas_mwh':               'Fossil Gas',
    'other_conventional_mwh':'Other Conventional',
    'hydro_pumped_mwh': 'Hydro Pumped Storage'
}

fct_gen = df[['timestamp_start'] + list(source_col_map.keys())].melt(
    id_vars='timestamp_start',
    var_name='col_name',
    value_name='generation_mwh'
)
fct_gen['source_name'] = fct_gen['col_name'].map(source_col_map)
fct_gen = fct_gen.merge(
    dim_energy_source[['source_id','source_name']], on='source_name')
fct_gen = fct_gen.merge(
    dim_time[['time_id','timestamp_start']], on='timestamp_start')
fct_gen = fct_gen[['time_id','source_id','generation_mwh']]
```

```python
import os

os.makedirs('data/processed', exist_ok=True)

# Save lại
fct_summary.to_csv('data/processed/fct_summary.csv', index=False)
fct_gen.to_csv('data/processed/fct_generation.csv', index=False)
dim_time.to_csv('data/processed/dim_time.csv', index=False)
dim_energy_source.to_csv('data/processed/dim_energy_source.csv', index=False)

print("Done saving CSV")
```

```python
from sqlalchemy import create_engine
import pandas as pd


engine = create_engine(
    'postgresql://postgres:'
)

# Test connection trước
try:
    with engine.connect() as conn:
        print("Connection OK ✓")
except Exception as e:
    print(f"Connection failed: {e}")
```

```python
print("Uploading dim_energy_source...")
dim_energy_source.to_sql('dim_energy_source', engine,
                          if_exists='replace', index=False)
print(f"✓ dim_energy_source: {len(dim_energy_source):,} rows")

print("Uploading dim_time...")
dim_time.to_sql('dim_time', engine,
                if_exists='replace', index=False)
print(f"✓ dim_time: {len(dim_time):,} rows")

print("Uploading fct_summary...")
fct_summary.to_sql('fct_summary', engine,
                   if_exists='replace', index=False,
                   chunksize=10000)
print(f"✓ fct_summary: {len(fct_summary):,} rows")

print("\n All done!")
```
